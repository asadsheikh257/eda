```python
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='right')
```

```python
final_df.head(2)
```

```python
# to check data type 
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# Pie Chart 
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

### Observation: Maximum numbers of records are from India and USA after then UK.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observation:
1. Rating between 4.5 to 4.9 ----> Excellent
2. Rating between 4.0 to 4.4 ----> Very Good
3. Rating between 3.5 to 3.9 ----> Good
4. Rating between 2.5 to 3.4 ----> Average
5. Rating between 1.8 to 2.4 ----> Poor
6. Rating is equal to 0.0 Means not rated

```python
ratings.head()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings)
plt.show()
```

```python
color = ['black','red','orange','yellow','green','green']
plt.figure(figsize=(16,8))
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette=color)
plt.show()
```

### Observation: 
1. Not Rated `Black` is very High
2. Maximum number of Ratings are between 2.6 to 4.3

```python
# countplot 
sns.countplot(x='Rating color', data=ratings, palette=color)
plt.show()
```
